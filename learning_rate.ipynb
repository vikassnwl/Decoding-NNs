```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
from utils import set_global_seed
```

```python
X = np.array([[1], [2], [3], [4]])
y = np.array([1, 2, 3, 4])
```

```python
def train_model(learning_rate, epochs):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(1,)),
        tf.keras.layers.Dense(1)
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss="mse")
    history = model.fit(X, y, epochs=epochs, verbose=0)
    return history
```

```python
for learning_rate in [0.1, 0.01, 0.005]:
    set_global_seed(42)
    history = train_model(learning_rate, epochs=50)
    losses = history.history["loss"]
    plt.plot(range(len(losses)), losses, label=learning_rate)
plt.title("Loss curves for different learning rates")
plt.legend()
plt.show()
```

**Observations:**
- It can be clearly seen in the plot above that if the learning rate is high enough then there is an oscillation in the loss curve which may introduce difficulty for model to learn properly.
- For learning rate 0.01, the curve is smoother than the learning rate 0.1.
- The smoothest curve is when learning rate is 0.005.

**Suggestions:**
- Choose an appropriate value for learning rate.
- A high value for learning rate may introduce oscillation during optimization which can prevent model to achieve minimum loss.
- A very low value for learning rate reduces the loss gradually but it can take too long to reach the minimum loss.
- Choose a value which is neither too high nor too low so that model could achieve minimum loss without oscillation in less time.
